# Columns as Lists

```python
sno = [1,2,3,4]
name = ['anu','alice','ken','bob']
age = [29, 52,48,25]
score = [500.25, 428.8, 562, 897.09]
languages = [('Python','Java'),('C', 'C++'),('Go'),('HTML', 'CSS')]
```

# Store Column by Column

```python
student = [sno, name, age, score, languages]
print(student)
```

# Store Row by Row

```python
student = [(1,'anu',29,500.29,('Python','Java')),
          (2,'alice',52,428.8,('C','C++')),
            (3,'ken',48,562,('Go')),
               (4,'bob',25,897.09,('HTML','CSS'))           
          ]
print(student)
```

# Store Row by Row - Better Option

```python
student = list(zip(sno, name, age, score, languages))
print(student)
```
